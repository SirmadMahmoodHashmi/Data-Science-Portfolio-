### **Simple Graph Plotting**

````python
#importing numpy - It is the fundamental package for scientific computing in Python
#reason for using numpy is to create arrays and perform some mathematical functions to do visualization.
import numpy as np
````

````python
#linspace: 
#Return: evenly spaced numbers over a specified interval.
x = np.linspace(0,10,100)
````

````python
#plotting the grap - plot() is used to plot a grap
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
````

````python
#plotting a the grap - using customized line shapes 
#available line styles - {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
plt.plot(x,np.sin(x), '-.')
plt.plot(x,np.cos(x), '--')
````

### **Sub Graph Plotting**

Blow code will show you how to subplots the graph?

````python
#creating first panel and setting its axis
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

#creating second panel and setting its axisplt.subplot(2,1,1)
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
````

## **Object-oriented interface**
It is most useful when dealing with complicated graph plotting scenarios, where you want more control over your figures. It will help you explicit control over you figure axes objects such as labels etc.  

````python
#plotting the grap with OOI
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x,np.cos(x))
````

### **lines color and style**

In this example, the liming axis of the plotting is also displayed.

````python
fig, ax = plt.subplots()
plt.plot(x, np.sin(x-0), color='orange', linestyle='solid', label = 'sin(x-0)') # Color by name
plt.plot(x, np.sin(x-1), color='g', linestyle='dashed', label = 'sin(x-1)') # short color code (rgbcmyk)
plt.plot(x, np.sin(x-2), color='0.85', linestyle='dashdot', label = 'sin(x-2)') # grayscale between 0 to 1
plt.plot(x, np.sin(x-3), color='#B8860B', linestyle='dotted', label = 'sin(x-3)') # hex code (RRGGBB from 00 to ff)
plt.plot(x, np.sin(x-4), color=(1.0, 0.0, 0.0), linestyle='-', label = 'sin(x-4)') # rbg tuple, value 0 and 1
plt.plot(x, np.sin(x-5), color='DeepPink', linestyle=':', label = 'sin(x-5)') # all htm color names supported
#setting the x limit of the plot
plt.xlim(-1,11)
##setting the y limit of the plot
plt.ylim(-1.5,1.5)
leg = ax.legend()
````

### **Histogram**

````python
#create a small dataset
data = np.random.randn(500)
fig_1 = plt.figure(1)
plt.hist(data, bins=30, alpha =1, color='steelblue')
fig_2 = plt.figure(2)
plt.hist(data, bins=30, histtype='stepfilled',alpha =1, color='steelblue')
plt.show()
````

````python
def f(x,y):
  return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x,y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
````
